## **Preproccessing**

### **Marked the rows containing EMOJIS**

```python
import re

def contains_emoji(text):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
    return emoji_pattern.search(text) is not None
```

```python
df_count = df[['è¯„è®ºå†…å®¹','è¯„è®ºæ—¶é—´']]
df_count['è¯„è®ºå†…å®¹'] = df_count['è¯„è®ºå†…å®¹'].astype(str)
df_count['è¯„è®ºæ—¶é—´'] = pd.to_datetime(df_count['è¯„è®ºæ—¶é—´'])
df_count['Year_Month'] = df_count['è¯„è®ºæ—¶é—´'].dt.strftime('%Y-%m')
df_count['count'] = df_count['è¯„è®ºå†…å®¹'].apply(lambda x: len(x))

df_count['contains_emoji'] = df_count['è¯„è®ºå†…å®¹'].apply(contains_emoji)
df_count.drop_duplicates(subset='è¯„è®ºå†…å®¹')
df_count
```

### **Split database on containing emojis**

```python
df_count_emoji = df_count[df_count['contains_emoji']==True]
df_count_no_emoji = df_count[df_count['contains_emoji']==False]
display(df_count)
display(df_count_emoji)
```

## **Sampling and saving**  
**Not containing emojis**  
First, divide it into layers based on the month.  
Then, evenly divide it into five layers from most to least words. Each layer should be randomly sampled at a rate of 0.011%.  
In order to maintain the distribution of words.    
  **Containing Emojis**  
Group by months and then samplling at a rate of 0.011%.  

```python
df_count = df
df_count['è¯„è®ºå†…å®¹'] = df_count['è¯„è®ºå†…å®¹'].astype(str)
df_count['è¯„è®ºæ—¶é—´'] = pd.to_datetime(df_count['è¯„è®ºæ—¶é—´'])
df_count['Year_Month'] = df_count['è¯„è®ºæ—¶é—´'].dt.strftime('%Y-%m')
df_count['count'] = df_count['è¯„è®ºå†…å®¹'].apply(lambda x: len(x))

df_count['contains_emoji'] = df_count['è¯„è®ºå†…å®¹'].apply(contains_emoji)
df_count.drop_duplicates(subset='è¯„è®ºå†…å®¹')
df_count
    
grouped = df_count_no_emoji.groupby('Year_Month', as_index=True, group_keys=True)
sampled_df = grouped.apply(lambda x:x.sort_values('count',ascending = False))
sampled_df['quantile'] = grouped.apply(lambda x: pd.qcut(x['count'], 5, labels=False))
sampled_df = sampled_df.apply(lambda x:x.reset_index(drop=True))
sampled_df = sampled_df.groupby(['Year_Month', 'quantile'], as_index=True, group_keys=True).apply(lambda x: x.sample(frac = 0.0011))
sampled_df = sampled_df.apply(lambda x:x.reset_index(drop=True))
sampled_df = sampled_df.drop(['è¯„è®ºæ—¶é—´','count','contains_emoji','quantile'], axis=1)

grouped_emoji = df_count_emoji.groupby('Year_Month', as_index=True, group_keys=True)
sampled_df_emoji = grouped_emoji.apply(lambda x:x.sample(frac = 0.0011))
sampled_df_emoji = sampled_df_emoji.apply(lambda x:x.reset_index(drop=True))
sampled_df_emoji = sampled_df_emoji.drop(['è¯„è®ºæ—¶é—´','count','contains_emoji'], axis=1)

sample = pd.concat([sampled_df, sampled_df_emoji],axis=0)
sample = sample.apply(lambda x:x.reset_index(drop=True))
sample = sample.sample(frac=1).reset_index()
sample['Whailing Wall'] = 0
sample['Tree Hole'] = 0
sample['Not related'] = 0
sample.to_csv('dataset/LiwenliangSample.csv', index=False, encoding='utf-8-sig')
display(sample)
```
