以下是对照组的树洞及哭墙的各账号分词数量，代码结构及思路与实验组相似

```python
import pandas as pd
```

```python
listx = ["红烧土豆叶","鹿道森","离灯_冬眠mode关闭失败","醒醒我们回家了-走饭小号","头发有点儿杵脖子","旅行的孤独风","玫瑰机关枪","十里清禾w","刘学州a","鸡蛋姬","走饭"]
```

哭墙的各账号分词数量统计，具体思路可见本路径readme

```python
for i in listx:

    df_words_all_explode_count = pd.read_csv(f'word_counts_csv/{i}_AllWords_count.csv',encoding="GBK")

    df_words_unique_tree_hole_explode_count = df_words_all_explode_count[['Words','Tree Hole']]
    df_words_unique_tree_hole_explode_count = df_words_unique_tree_hole_explode_count.sort_values(by=['Tree Hole'], ascending=False)
    df_words_unique_tree_hole_explode_count = df_words_unique_tree_hole_explode_count.reset_index(drop=True)

    df_words_unique_wailing_wall_explode_count = df_words_all_explode_count[['Words','Wailing Wall']]
    df_words_unique_wailing_wall_explode_count = df_words_unique_wailing_wall_explode_count.sort_values(by=['Wailing Wall'], ascending=False)
    df_words_unique_wailing_wall_explode_count = df_words_unique_wailing_wall_explode_count.reset_index(drop=True)

    df_words_both_explode_count = df_words_all_explode_count[['Words','Both']]
    df_words_both_explode_count = df_words_both_explode_count.sort_values(by=['Both'], ascending=False)
    df_words_both_explode_count = df_words_both_explode_count.reset_index(drop=True)

    df_words_all_explode_count_difference = df_words_all_explode_count[df_words_all_explode_count['All'] > 10]

    df_words_all_explode_count_difference['Difference'] = (abs(df_words_all_explode_count_difference['Wailing Wall'] - df_words_all_explode_count_difference['Tree Hole']) / (df_words_all_explode_count_difference['Wailing Wall'] + df_words_all_explode_count_difference['Tree Hole'])) * 100
    df_words_all_explode_count_difference = df_words_all_explode_count_difference.sort_values(by=['Difference'], ascending=False)

    words_count_unique_wailing_wall = df_words_all_explode_count_difference[df_words_all_explode_count_difference['Wailing Wall'] > df_words_all_explode_count_difference['Tree Hole']]
    words_count_unique_tree_hole = df_words_all_explode_count_difference[df_words_all_explode_count_difference['Tree Hole'] > df_words_all_explode_count_difference['Wailing Wall']]
    
    rows_to_keep1 = int(words_count_unique_tree_hole.shape[0] * 0.75)
    tree_hole_subset = words_count_unique_tree_hole.iloc[:rows_to_keep1]

    tree_hole_subset.insert(0, "ID", i)

    tree_hole_subset.to_csv(f'every_words_difference_csv/tree_hole/{i}_tree_hole.csv', index=False)
```
